```python
import pypsa
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
network = pypsa.Network()

# Ajouter des bus avec des positions géographiques (coordonnées approximatives)
network.add("Bus", "Paris", x=2.3522, y=48.8566)       # Paris
network.add("Bus", "Lyon", x=4.8357, y=45.7640)       # Lyon
network.add("Bus", "Marseille", x=5.3698, y=43.2965)  # Marseille
network.add("Bus", "Toulouse", x=1.4442, y=43.6047)   # Toulouse
network.add("Bus", "Lille", x=3.0573, y=50.6292)      # Lille

# Ajouter des lignes reliant les bus
network.add("Line", "line1", bus0="Paris", bus1="Lyon", x=0.1, r=0.01)
network.add("Line", "line2", bus0="Lyon", bus1="Marseille", x=0.1, r=0.01)
network.add("Line", "line3", bus0="Paris", bus1="Lille", x=0.1, r=0.01)
network.add("Line", "line4", bus0="Toulouse", bus1="Marseille", x=0.1, r=0.01)

# Ajouter un générateur et une charge
network.add("Generator", "gen1", bus="Paris", p_nom=100, marginal_cost=50)
network.add("Load", "load1", bus="Marseille", p_set=50)

network.plot(color_geomap=True, projection=ccrs.EqualEarth(), margin=1)
```
